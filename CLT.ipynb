# Доказательство центральной предельной теоремы

В работе рассматривается непрерывное двухпараметрическое распределение Парето (Брэдфорда) и эмпирически доказывается истинность классическй центральной предельной теоремы на его примере.

Сгенерируем случайную выборку размера 1000, построим её гистограмму и график теоретической плотности распределения:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
b = 50
random_sample = sts.pareto.rvs(b, size = 1000)
```

```python
plt.hist(random_sample, bins = 30,  color = 'green', normed=True)
plt.title('Random sample (size = 1000)')
plt.ylabel('Number of cases')
plt.xlabel('x')
x = np.linspace(1, 1.14, 1000)
pdf = sts.pareto(b).pdf(x)
plt.plot(x, pdf, label='Parent distribution', color = 'red', alpha=5)
plt.legend()
plt.show()
```

Сгенерируем 1000 выборок объёма 5 и построим гистограммы распределений их выборочных средних:

```python
means_5 = []
for i in range(1000):
    b = 50
    random_sample_5 = sts.pareto.rvs(b, size = 5)
    means_5.append(random_sample_5.mean())
    
plt.hist(means_5, bins = 30,  color = 'green')
plt.ylabel('Number of means (size = 5)')
plt.xlabel('Mean')
plt.title('1000 samples of n = 5')
plt.show()
```

Сгенерируем 1000 выборок объёма 10 и построим гистограммы распределений их выборочных средних:

```python
means_10 = []
for i in range(1000):
    b = 50
    random_sample_10 = sts.pareto.rvs(b, size = 10)
    means_10.append(random_sample_10.mean())
plt.hist(means_10, bins = 30,  color = 'green')
plt.ylabel('Number of means (size = 10)')
plt.xlabel('Mean')
plt.title('1000 samples of n = 10')
plt.show()
```

Сгенерируем 1000 выборок объёма 50 и построим гистограммы распределений их выборочных средних:

```python
means_50 = []
for i in range(1000):
    b = 50
    random_sample_50 = sts.pareto.rvs(b, size = 50)
    means_50.append(random_sample_50.mean())
plt.hist(means_50, bins = 30,  color = 'green')
plt.ylabel('Number of means (size = 50)')
plt.xlabel('Mean')
plt.title('1000 samples of n = 50')
plt.show()
```

Приблизим распределения выборочных средних нормальными при разных n, то есть проверим работу центральной предельной теоремы:

```python
mean = sts.pareto(b).mean()
variance = sts.pareto(b).var()
```

```python
plt.hist(means_5, 30, normed=True, color='green')
stand_error = np.sqrt(variance)/np.sqrt(5)
norm = sts.norm(mean, stand_error) 
plt.plot(x, norm.pdf(x), label='Parent distribution', color = 'red', alpha = 5)
plt.title('CLT for n = 5')
plt.xlabel('Mean')
plt.ylabel('Number of means (size = 5)')
plt.legend()
plt.show()
```

```python
plt.hist(means_5, 30, normed=True, color='green')
stand_error = np.sqrt(variance)/np.sqrt(10)
norm = sts.norm(mean, stand_error) 
plt.plot(x, norm.pdf(x), label='Parent distribution', color = 'red', alpha = 5)
plt.title('CLT for n = 10')
plt.xlabel('Mean')
plt.ylabel('Number of means (size = 10)')
plt.legend()
plt.show()
```

```python
plt.hist(means_50, 30, normed=True, color='green')
stand_error = np.sqrt(variance)/np.sqrt(50)
norm = sts.norm(mean, stand_error) 
plt.plot(x, norm.pdf(x), label='Parent distribution', color = 'red', alpha = 5)
plt.title('CLT for n = 50')
plt.xlabel('Mean')
plt.ylabel('Number of means (size = 50)')
plt.legend()
plt.show()
```
